```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv('insurance_data.csv', index_col = None)
```

```python
df
```

```python
plt.scatter(df.age, df.bought_insurance, marker= '+', color= 'red')
plt.show()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test =  train_test_split (df[['age']], df[['bought_insurance']], test_size= 0.1, random_state= 42)
```

```python
x_train
```

```python
x_test
```

```python
# creating model 
from sklearn import linear_model
```

```python
reg= linear_model.LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
x_test
```

```python
y_test
```

```python
# Now predicting values based on test data
reg.predict(x_test)

# answer is correct
```

```python
# Evaluating accuracy of model 
# Model is perfect as data size is small
reg.score(x_test, y_test)
```
